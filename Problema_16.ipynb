# **PROBLEMA** **16**

Dos personas han quedado en una cita en un determinado lugar. Ambas tienen la misma probabilidad de llegar al lugar de la cita en el intervalo de tiempo [0, 30]. 
Determínese la probabilidad de que el tiempo que una persona espere a la otra sea como máximo 10. 


###**Solución analítica**

El tiempo de espera tendrá una distribución uniforme en el intervalo [0, 30], ya que la otra persona puede llegar en cualquier momento dentro de ese intervalo.
Como la probabilidad se puede expresar como un área bajo la curva de la distribución uniforme, graficamos:

343270645_3062696644040231_947841977606989_n.jpg

Ahora, calculamos la probabilidad:  

$P(X-Y \leq 10)=\frac{Área\ sombreada}{Área\ total}=\frac{500}{900}=\frac{5}{9}$  
$\therefore$ la probabilidad de que el tiempo máximo de espera sea de 10 minutos, es del 55.56%.

### **1. Identifique las variables**

Sea $x_1$=Tiempo de espera de la primera persona.  
Sea $x_2$=Tiempo de espera de la segunda persona.

### **2. Determine la distribución de probabilidad**

Se usará la distribución uniforme continua, ya que se especifica que las dos personas tienen la misma probabilidad de llegar al lugar de la cita en el intervalo de tiempo [0, 30].

### **3. Modele las variables aleatorias**

$x_1∼U(0,30)$  
$x_2∼U(0,30)$

### **4. Defina el modelo del sistema y los objetivos de la simulación**

El sistema en este caso es la cita entre dos personas en un determinado lugar. Las variables del sistema son los tiempos de llegada de cada persona al lugar de la cita. Se asume que estos tiempos de llegada siguen una distribución uniforme en el intervalo [0, 30].  

El objetivo de la simulación es obtener una estimación de la probabilidad de que el tiempo de espera sea como máximo 10 utilizando el modelo del sistema y generando múltiples escenarios de llegada de las dos personas.

```python
import numpy as np
import random 
import matplotlib.pyplot as plt 
```

```python
def llegada():
  llega1=0
  llega2=0
  p1=random.randrange(0, 30, 1) #Generar tiempos de llegada aleatorios
  p2=random.randrange(0, 30, 1)
  if p1<p2: #Determinar quien llega primero
    llega1=p1
    llega2=p2
  else:
    llega1=p2
    llega2=p1
  espera=llega2-llega1 #Determinar el tiempo de espera
  return(espera) 
```

```python
def simulacion(n):
  L=[]
  c=0
  for i in range(n):
    if llegada()<=10: #Se verifica si el tiempo de espera es menor o igual a 10
      c=c+1
      L.append(c/(i+1)) 
  return(L)
```

###**5. Diseñe el experimento**

```python
x=simulacion(50) #Asiganr una variable para todas las simulaciones
```

```python
plt.plot(x) #Gráfica
```

```python
np.mean(x) 
```

```python
def modelo(n):
    lista=[]
    for i in range(n):
        s = simulacion(100)
        lista.append(np.mean(s)) 
    return lista
```

###**6. Repita el experimento $n$ veces**

```python
Y=modelo(50) #Repetir modelo un determinado número de veces más
```

```python
print(np.mean(Y), np.std(Y))
```

```python
plt.hist(Y) #Gráfica del histograma
plt.show()
```

###**7. Hallar el intervalo de confianza**

```python
print("El intervalo de confianza es: ",
np.mean(Y)-(1.96)*((np.var(Y))/np.sqrt(50)), "≤ X̅ ≤",
np.mean(Y)+(1.96)*((np.var(Y))/np.sqrt(50)))
```
